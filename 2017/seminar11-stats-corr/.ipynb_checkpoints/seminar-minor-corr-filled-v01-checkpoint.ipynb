## Foodmart product sales 

Загрузите данные `foodmard.sales.tsv` и `foodmart.products.tsv`. Вам необходимо выявить взаимосвязь в продажах продуктов по всем магазинам.

```python
sales = pd.read_csv('foodmart.sales.tsv', sep = '\t', header = 0, parse_dates = [2])
```

```python
sales.head()
```

```python
products = pd.read_csv('foodmart.products.tsv', sep = '\t', header = 0)
```

```python
products.head()
```

```python
sales = sales.merge(products[['product_id', 'product_name']], 
                    on = ['product_id'], how = 'inner')
```

```python
sales.head()
```

## Взаимное влияние продуктов друг на друга

Составьте таблицу вида  (Дата, магазин) - (Товар), так, чтобы в соответствующей ячейке было указано количество единиц товара купленного в соответствующий день в соответствующем магазине.

```python
pivot_sales = sales.pivot_table(index=['date', 'store_id'], 
                                columns='product_name',
                                values='sales', 
                                aggfunc='sum',
                                fill_value=0)
```

```python
pivot_sales.shape
```

```python
pivot_sales.head()
```

Посчитайте корреляцию

```python
sales_corr = pivot_sales.corr()
```

```python
import seaborn as sns
```

```python
# sns.heatmap(sales_corr)
```

```python
sales_corr.head()
```

```python
idx = np.triu_indices_from(sales_corr.values)
```

```python
corr = sales_corr.values
```

```python
corr[idx] = np.nan
```

```python
sales_corr.iloc[:, :] = corr
```

Найдите товары с наибольшими значениями корреляции

```python
sales_corr_unstack = sales_corr.unstack(level=1).to_frame()
sales_corr_unstack.columns = ['correlation']
sales_corr_unstack.index.names = ['product1', 'product2']
```

```python
sales_corr_unstack = sales_corr_unstack.dropna()
```

```python
idx = sales_corr_unstack.index.levels[0].values == sales_corr_unstack.index.levels[1].values
```

```python
idx = sales_corr_unstack.index.get_level_values(0)!= sales_corr_unstack.index.get_level_values(1)
```

```python
sales_corr_unstack = sales_corr_unstack.loc[idx, :]
```

```python
sales_corr_unstack = sales_corr_unstack.sort_index(by='correlation')
```
